```python
#!/usr/bin/env python3
"""
Bç«™è§†é¢‘æ•°æ®åˆ†æ - ç»ˆæç²¾ç®€ç‰ˆ
è¾“å…¥BVå·ï¼Œè·å–ä¸‰è¿æ•°æ®å¹¶ç”ŸæˆæŸ±çŠ¶å›¾
"""

import asyncio
import matplotlib.pyplot as plt
from bilibili_api import video

class BilibiliQuickAnalyzer:
    def __init__(self):
        plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei']
        plt.rcParams['axes.unicode_minus'] = False
    
    async def get_video_data(self, bvid):
        """è·å–è§†é¢‘æ•°æ®"""
        try:
            v = video.Video(bvid=bvid)
            info = await v.get_info()
            return info
        except Exception as e:
            print(f"âŒ è·å–è§†é¢‘å¤±è´¥: {e}")
            return None
    
    def analyze_and_plot(self, bvid):
        """åˆ†æå¹¶ç»˜å›¾ - ä¸»å‡½æ•°"""
        print(f"ğŸ¯ æ­£åœ¨åˆ†æè§†é¢‘: {bvid}")
        
        # è·å–æ•°æ®
        info = asyncio.run(self.get_video_data(bvid))
        if not info:
            return
        
        # æå–å…³é”®æ•°æ®
        stat = info['stat']
        data = {
            'æ ‡é¢˜': info['title'][:20] + '...' if len(info['title']) > 20 else info['title'],
            'æ’­æ”¾é‡': stat['view'],
            'ç‚¹èµæ•°': stat['like'], 
            'æŠ•å¸æ•°': stat['coin'],
            'æ”¶è—æ•°': stat['favorite']
        }
        
        # æ‰“å°ç»“æœ
        print(f"\nğŸ“Š åˆ†æç»“æœ:")
        print(f"æ ‡é¢˜: {info['title']}")
        print(f"UPä¸»: {info['owner']['name']}")
        print(f"æ’­æ”¾é‡: {data['æ’­æ”¾é‡']:,}")
        print(f"ç‚¹èµæ•°: {data['ç‚¹èµæ•°']:,}")
        print(f"æŠ•å¸æ•°: {data['æŠ•å¸æ•°']:,}") 
        print(f"æ”¶è—æ•°: {data['æ”¶è—æ•°']:,}")
        
        # è®¡ç®—æ¯”ç‡
        total = data['ç‚¹èµæ•°'] + data['æŠ•å¸æ•°'] + data['æ”¶è—æ•°']
        print(f"ä¸‰è¿æ€»æ•°: {total:,}")
        print(f"ç‚¹èµç‡: {data['ç‚¹èµæ•°']/data['æ’­æ”¾é‡']:.2%}")
        print(f"æŠ•å¸ç‡: {data['æŠ•å¸æ•°']/data['æ’­æ”¾é‡']:.2%}")
        print(f"æ”¶è—ç‡: {data['æ”¶è—æ•°']/data['æ’­æ”¾é‡']:.2%}")
        
        # ç”ŸæˆæŸ±çŠ¶å›¾
        self.generate_bar_chart(data, bvid)
    
    def generate_bar_chart(self, data, bvid):
        """ç”Ÿæˆä¸‰è¿æ•°æ®æŸ±çŠ¶å›¾"""
        labels = ['æ’­æ”¾é‡', 'ç‚¹èµæ•°', 'æŠ•å¸æ•°', 'æ”¶è—æ•°']
        values = [data['æ’­æ”¾é‡'], data['ç‚¹èµæ•°'], data['æŠ•å¸æ•°'], data['æ”¶è—æ•°']]
        colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
        
        plt.figure(figsize=(10, 6))
        bars = plt.bar(labels, values, color=colors, alpha=0.7)
        
        # è®¾ç½®å›¾è¡¨æ ·å¼
        plt.title(f'Bç«™è§†é¢‘æ•°æ®åˆ†æ\n{data["æ ‡é¢˜"]}\nBV: {bvid}', fontsize=14)
        plt.ylabel('æ•°é‡')
        
        # åœ¨æŸ±å­ä¸Šæ˜¾ç¤ºæ•°å€¼
        for bar, value in zip(bars, values):
            plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(values)*0.01, 
                    f'{value:,}', ha='center', va='bottom', fontweight='bold')
        
        plt.grid(axis='y', alpha=0.3)
        plt.tight_layout()
        plt.show()

# ä½¿ç”¨ç¤ºä¾‹
if __name__ == "__main__":
    analyzer = BilibiliQuickAnalyzer()
    
    # ç›´æ¥åœ¨è¿™é‡Œè¾“å…¥BVå·ï¼
    bv_id = "BV1cSnuzYE88"  # ğŸ”¥ ä¿®æ”¹è¿™é‡Œçš„BVå·ï¼
    
    analyzer.analyze_and_plot(bv_id)
```

```python
# åˆ›å»º core.py å†…å®¹
core_code = '''
"""
Bç«™è§†é¢‘åˆ†ææ ¸å¿ƒæ¨¡å—
"""
import asyncio
import time
from bilibili_api import video

class BilibiliAnalyzer:
    """Bç«™è§†é¢‘åˆ†æå™¨"""
    
    def __init__(self):
        self.sequence = []
        self.start_time = time.time()
    
    def _checkpoint(self, step_name):
        """å†…éƒ¨æ–­ç‚¹æ£€æŸ¥"""
        timestamp = time.time() - self.start_time
        self.sequence.append(1)
        print(f"â° [{timestamp:.3f}s] {step_name}: 1")
        return timestamp
    
    async def analyze_video(self, bv_id):
        """åˆ†æè§†é¢‘æ•°æ®"""
        try:
            # åˆ›å»ºè§†é¢‘å¯¹è±¡
            self._checkpoint("åˆ›å»ºè§†é¢‘å¯¹è±¡")
            v = video.Video(bvid=bv_id)
            
            # è·å–è§†é¢‘ä¿¡æ¯
            self._checkpoint("è·å–APIæ•°æ®")
            info = await v.get_info()
            
            # æå–æ•°æ®
            self._checkpoint("è§£ææ•°æ®")
            title = info['title']
            up_name = info['owner']['name']
            stat = info['stat']
            
            # æ•´ç†ç»“æœ
            result = {
                'bv_id': bv_id,
                'title': title,
                'up_name': up_name,
                'view': stat['view'],
                'like': stat['like'],
                'coin': stat['coin'],
                'favorite': stat['favorite'],
                'share': stat['share'],
                'reply': stat['reply'],
                'sequence': self.sequence.copy(),
                'total_time': time.time() - self.start_time
            }
            
            # è®¡ç®—æ¯”ç‡
            result['like_rate'] = result['like'] / result['view']
            result['coin_rate'] = result['coin'] / result['view']
            result['favorite_rate'] = result['favorite'] / result['view']
            
            self._checkpoint("åˆ†æå®Œæˆ")
            return result
            
        except Exception as e:
            print(f"âŒ åˆ†æå¤±è´¥: {e}")
            return None

class InteractiveTester:
    """äº¤äº’å¼æµ‹è¯•å™¨"""
    
    def __init__(self):
        self.analyzer = BilibiliAnalyzer()
    
    async def run_interactive_test(self):
        """è¿è¡Œäº¤äº’å¼æµ‹è¯•"""
        while True:
            print("\\n" + "="*50)
            bv_id = input("è¯·è¾“å…¥BVå· (è¾“å…¥ 'q' é€€å‡º): ").strip()
            
            if bv_id.lower() == 'q':
                print("æµ‹è¯•ç»“æŸï¼")
                break
                
            if not bv_id.startswith('BV'):
                print("âŒ è¯·è¾“å…¥æœ‰æ•ˆçš„BVå·")
                continue
            
            print(f"ğŸ¯ æ­£åœ¨åˆ†æ: {bv_id}")
            result = await self.analyzer.analyze_video(bv_id)
            
            if result:
                self._display_result(result)
            else:
                print("âŒ åˆ†æå¤±è´¥")
    
    def _display_result(self, result):
        """æ˜¾ç¤ºåˆ†æç»“æœ"""
        print(f"\\nâœ… åˆ†ææˆåŠŸ!")
        print(f"ğŸ“º æ ‡é¢˜: {result['title']}")
        print(f"ğŸ‘¤ UPä¸»: {result['up_name']}")
        print(f"ğŸ“Š æ’­æ”¾é‡: {result['view']:,}")
        print(f"ğŸ‘ ç‚¹èµ: {result['like']:,} ({result['like_rate']:.2%})")
        print(f"ğŸª™ æŠ•å¸: {result['coin']:,} ({result['coin_rate']:.2%})")
        print(f"â­ æ”¶è—: {result['favorite']:,} ({result['favorite_rate']:.2%})")
        print(f"ğŸ”¢ æ‰§è¡Œåºåˆ—: {''.join(str(x) for x in result['sequence'])}")
        print(f"â±ï¸  æ€»è€—æ—¶: {result['total_time']:.3f}s")
'''

with open('bilibili_analyzer/src/core.py', 'w', encoding='utf-8') as f:
    f.write(core_code)

print("âœ… core.py åˆ›å»ºå®Œæˆ!")
```
